<h1>Optical Character Recognition using EasyOCR python library </h1>

<h3>Step1: Import python libraries </h3>

```python
import easyocr
import cv2
from matplotlib import pyplot as plt
import numpy as np
from IPython.display import Image
import csv
import re
import uuid
import pandas as pd
```

<h3>Step2: Open the image to be processed</h3>

```python
IMAGE_PATH = "D://OCR project//processed_tripsheet//5.jpg"
print(IMAGE_PATH)
IMAGE_PATH
Image(filename="D://OCR project//processed_tripsheet//5.jpg")
```

<h3>Step3: Instantiate <i> Reader </i> object and pass the opened image </h3>

```python
reader = easyocr.Reader(['en'], gpu=False)
result = reader.readtext(IMAGE_PATH)
result
```

<h3>Step4: Visualise the obtained results </h3>

```python
img = cv2.imread(IMAGE_PATH)
for detection in result:
    top_left = tuple([int(val) for val in detection[0][0]])
    bottom_right = tuple([int(val) for val in detection[0][2]])
    text = detection[1]
    font = cv2.FONT_HERSHEY_SIMPLEX
    img = cv2.rectangle(img, top_left, bottom_right, (0,255,0), 5)
    img = cv2.putText(img, text, top_left, font, 1, (255,255,255), 2, cv2.LINE_AA)
    
plt.figure(figsize = (20,20))
plt.imshow(img)
plt.show()
```
